# Q2. 

In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. 

In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. 

Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
Brand = []
Title = []
Rating = []
No_of_rating = []
Price = []
Returns =[]
Expected_delivery = []
Availability = []
Details = []
Url = []
```

```python
# creating individual functions for each column

def a():
    try:
        brand_name = driver.find_element_by_xpath("//div[@id='bylineInfo_feature_div']//div//a")
        Brand.append(brand_name.text.replace('Brand: ',''))
    except NoSuchElementException as n:
        Brand.append('-')
        print( 'Exception:', n)
    
def b():
    try:
        product_name = driver.find_element_by_xpath("//h1[@id = 'title']//span")
        Title.append(product_name.text)
    except NoSuchElementException as n:
        Title.append('-')
        print( 'Exception:', n)
def c():
    try:
        rating = driver.find_element_by_xpath("//div[@class='a-fixed-left-grid-col aok-align-center a-col-right']//div//span//span")
        Rating.append(rating.text)
    except NoSuchElementException as n:
        Rating.append('-')
        print( 'Exception:', n)
    
def d():
    try:
        no_of_rating = driver.find_element_by_xpath("//div[@data-hook='total-review-count']//span")
        No_of_rating.append(no_of_rating.text)
    except NoSuchElementException as n:
        No_of_rating.append('-')
        print( 'Exception:', n)
    
def e():
    try:
        price = driver.find_element_by_xpath("//tr[@id='priceblock_ourprice_row']//td[@class='a-span12']")
        Price.append(price.text)
    except NoSuchElementException as n:
        Price.append('-')
        print( 'Exception:', n)
    
def f():
    try:
        exchange = driver.find_element_by_xpath("//div[@id='icon-farm']//div[@id='RETURNS_POLICY']//div[@class='a-section a-spacing-none icon-content']")
        Returns.append(exchange.text)
    except NoSuchElementException as n:
        Returns.append('-')
        print( 'Exception:', n)
def g():
    try:
        delivery = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']//b")
        Expected_delivery.append(delivery.text)
    except NoSuchElementException as n:
        Expected_delivery.append('-')
        print( 'Exception:', n)
def h():
    try:
        availability = driver.find_element_by_xpath("//div[@id='availability_feature_div']//div//span")
        Availability.append(availability.text)
    except NoSuchElementException as n:
        Availability.append('-')
        print( 'Exception:', n)
def i():
    try:
        details = driver.find_element_by_xpath("//div[@id='productDescription']//p")
        Details.append(details.text)
    except NoSuchElementException as n:
        Details.append('-')
        print( 'Exception:', n)
    
def j():
    Url.append(driver.current_url)
```

```python
for k in (1,2,3): # since, we needed to scrap data from the first three pages
    data = driver.find_elements_by_xpath("//h2[@class = 'a-size-mini a-spacing-none a-color-base s-line-clamp-4']//a")
    
    for l in data:
        l.click()
    
        driver.switch_to.window(driver.window_handles[-1])
        time.sleep(2)
    
        a()
        b()
        c()
        d()
        e()
        f()
        g()
        h()
        i()
        j()
    
        driver.close()
        
        driver.switch_to.window(driver.window_handles[0])
    
        time.sleep(1)
    
    next_btn = driver.find_element_by_xpath("//li[@class = 'a-last']//a")
    next_btn.click()
    
    time.sleep(2)
    
    if k == 3: #creating the dataframe at last
        df = pd.DataFrame({'Brand':Brand, 'Title':Title, 'Rating':Rating, 'No_of_Rating':No_of_rating, 'Price':Price,
                          'Return_Policy':Returns, 'Expected_delivery': Expected_delivery, 
                          'Availability':Availability, 'Details':Details , 'Url':Url})
```

# Q3. 
Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.

```python
def image(): #defining a function to scrape images of anything which user inputs
    
    driver = webdriver.Chrome(r'C:\Users\AKHIL\Downloads\chromedriver_win32\chromedriver.exe')
    
    driver.maximize_window()
    
    driver.get('https://images.google.com/')
    
    search_title = driver.find_element_by_xpath("//input[@class='gLFyf gsfi']")
    
    search_title.send_keys(input('which images you want to get?')) #taking input from the user
    
    search_btn = driver.find_element_by_xpath("//button[@aria-label='Google Search']")
    
    search_btn.click()
    
    time.sleep(2)
    
    driver.switch_to.window(driver.window_handles[0])
    
    data = driver.find_elements_by_xpath("//div[@class = 'bRMDJf islir']//img")
    
    while len(data) <= 100:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        driver.switch_to.window(driver.window_handles[0])
        time.sleep(2)
        data = driver.find_elements_by_xpath("//div[@class = 'bRMDJf islir']//img")
        
    var = []
    for i in data:
        if len(var) < 100:
            var.append(i.get_attribute('src'))
            time.sleep(0.01)
    
    driver.close()
    
    return var
```

```python
Brand_name = []
Smartphone_name = []
Colour = []
RAM = []
Storage = []
Primary_camera = []
Secondary_camera = []
Display_size = []
Display_resolution = []
Processor = []
Processor_cores = []
Battery_capacity =  []
Price = []
Product_url = []
```

```python
def specs_data(): #defining a function to scrape the required details only
    driver.switch_to.window(driver.window_handles[-1])
    
    time.sleep(1)
    
    full_spec_btn = driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']")
    full_spec_btn.click()
    
    time.sleep(1)

    specs = driver.find_elements_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ']//div")

    for i in range(len(specs)):

        if specs[i].text == 'General':
            headings = driver.find_elements_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr//td[1]".format(i+1))
            for j in range(len(headings)):
                if headings[j].text == 'Color':
                    value = driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr[{}]//td[2]".format(i+1,j+1))
                    Colour.append(value.text)
                
                
        if specs[i].text == 'Display Features':
            headings = driver.find_elements_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr//td[1]".format(i+1))
            for j in range(len(headings)):
                if headings[j].text == 'Display Size':
                    value = driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr[{}]//td[2]".format(i+1,j+1))
                    Display_size.append(value.text)
                
                if headings[j].text == 'Resolution':
                    value = driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr[{}]//td[2]".format(i+1,j+1))
                    Display_resolution.append(value.text)
                
                
        if specs[i].text == 'Os & Processor Features':
            headings = driver.find_elements_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr//td[1]".format(i+1))
            for j in range(len(headings)):
                if headings[j].text == 'Processor Type':
                    value = driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr[{}]//td[2]".format(i+1,j+1))
                    Processor.append(value.text)
                
                if headings[j].text == 'Processor Core':
                    value = driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr[{}]//td[2]".format(i+1,j+1))
                    Processor_cores.append(value.text)
                
                
        if specs[i].text == 'Memory & Storage Features':
            headings = driver.find_elements_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr//td[1]".format(i+1))
            for j in range(len(headings)):
                if headings[j].text == 'Internal Storage':
                    value = driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr[{}]//td[2]".format(i+1,j+1))
                    Storage.append(value.text)
                
                if headings[j].text == 'RAM':
                    value = driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr[{}]//td[2]".format(i+1,j+1))
                    RAM.append(value.text)
    
        if specs[i].text == 'Camera Features':
            headings = driver.find_elements_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr//td[1]".format(i+1))
            for j in range(len(headings)):
                if headings[j].text == 'Primary Camera':
                    value = driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr[{}]//td[2]".format(i+1,j+1))
                    Primary_camera.append(value.text)
                
                if headings[j].text == 'Secondary Camera':
                    value = driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr[{}]//td[2]".format(i+1,j+1))
                    Secondary_camera.append(value.text)
                
                
        if specs[i].text == 'Battery & Power Features':
            headings = driver.find_elements_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr//td[1]".format(i+1))
            for j in range(len(headings)):
                if headings[j].text == 'Battery Capacity':
                    value = driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[@class='_3k-BhJ'][{}]//table//tr[{}]//td[2]".format(i+1,j+1))
                    Battery_capacity.append(value.text)
                
    name = driver.find_elements_by_xpath("//div[@class='_1MR4o5']//div[@class='_3GIHBu']")
    Brand_name.append(name[-2].text)
    
    title = driver.find_element_by_xpath("//h1[@class='yhB1nd']//span")
    Smartphone_name.append(title.text)
    
    rate = driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']")
    Price.append(rate.text)
    
    Product_url.append(driver.current_url)

    driver.close()

    driver.switch_to.window(driver.window_handles[0])
```

# Q5.
Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
driver = webdriver.Chrome(r'C:\Users\AKHIL\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
def lat_long (driver,city): #defining a function to take input from user and provide him with the coordinates
    
    driver.get("https://maps.google.com/")
    
    driver.maximize_window()
    
    search_address = driver.find_element_by_xpath("//input[@class='tactile-searchbox-input']")
    search_address.send_keys(city)
    
    search_btn = driver.find_element_by_id("searchbox-searchbutton")
    search_btn.click()
    
    driver.switch_to.window(driver.window_handles[0])
    
    time.sleep(5)
    
    url = driver.current_url #getting the url because it contains the coordinates in it
    
    return url.split('@')[1].split(',')[0:2]
```

```python
months  = driver.find_elements_by_xpath("//h2")
```

```python
def funding(): # creating a function to scrape funding details of specific month
    Mo_ye = input("Enter the month as below example \n 'July, 2018' \n " )
    
    df = pd.DataFrame({})
    for i in range(len(months)):
        Header_cols = []
        if months[i].text == Mo_ye:
            tables = driver.find_elements_by_xpath("//div[@class='wpb_text_column wpb_content_element  bs-vc-block'][1]//div[@class='dataTables_wrapper no-footer'][{}]//div[@class='dataTables_scroll']//div[1]//div//th".format(i))
            for t in tables:
                Header_cols.append(t.text)
            
            entries = driver.find_elements_by_xpath("//div[@class='wpb_text_column wpb_content_element  bs-vc-block'][1]//div[@class='dataTables_wrapper no-footer'][{}]//div[1]//option".format(i))
            entries[-1].click()
        
            for j in range(len(Header_cols)):
                var = []
            
                data = driver.find_elements_by_xpath("//div[@class='wpb_text_column wpb_content_element  bs-vc-block'][1]//div[@class='dataTables_wrapper no-footer'][{}]//div[@class='dataTables_scroll']//div[2]//tbody//tr//td[{}]".format(i,j+1))
            
                for k in range(len(data)):
                    var.append(data[k].text)
                    
                df[Header_cols[j]] = var
    
    return df
```

# Q7
Write a program to scrap all the available details of top 10 gaming laptops from digit.in

```python
driver = webdriver.Chrome(r'C:\Users\AKHIL\Downloads\chromedriver_win32\chromedriver.exe')
driver.get('https://www.digit.in/')

driver.maximize_window()
```

```python
top_10 = driver.find_element_by_xpath("//div[@class='menu']//ul//li[4]")
top_10.click()
driver.switch_to.window(driver.window_handles[0])
```

A pop up appears for notification permissions... make sure to close that pop up when you see it

```python
top_10_gam = driver.find_elements_by_xpath("//div[@class='listing_container']//ul//li")
title = input("Enter the text")

for i in top_10_gam:
    if i.text == title:
        i.click()
        
        driver.switch_to.window(driver.window_handles[0])
        
        break
```

```python
urls = driver.find_elements_by_xpath("//div[@data-cat='Top_Ten_En_Scroll']//a")
pages = [i.get_attribute('href') for i in urls]
```

```python
for i in range(len(pages)):
    
    driver.get(pages[i])
    
    driver.switch_to.window(driver.window_handles[0])
    
    d = [] 
    
    time.sleep(1)
    
    sidebar = driver.find_elements_by_xpath("//div[@class='tab']//a//button")
    
    var = []
    
    for j in range(len(sidebar)):
        
        sidebar[j].click()
        
        time.sleep(1)
        
        data = driver.find_elements_by_xpath("//div[@class='tabcontent']//div[{}]//div[@class='remaining-details']//tr".format(j+1))
        
        for k in data:
            
            var.append(k.text)
        
    x= [l.split(":") for l in var]
        
    cols = [l[0] for l in x]
        
    rows = [l[1] for l in x]
        
    d.append(rows)
    
    if i == 0:
        df = pd.DataFrame(data = d, columns = cols)
        
    else:
        df1 = pd.DataFrame(data = d, columns = cols)
        df = pd.concat([df,df1])

df = df.fillna('-') #fills all NaN values with '-'
df = df.reset_index().drop(columns = 'index') #resets the index and drops the column of old index

print(df)
```
